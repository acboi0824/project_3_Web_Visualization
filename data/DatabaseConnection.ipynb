```python
import pandas as pd

cleaned_pet_df = pd.read_csv('pet_data_cleaned.csv')
states_csv = pd.read_csv('states_coordinates.csv')
# print(cleaned_pet_df)
# print(states_csv)
```

```python
sr = cleaned_pet_df['primary']
print(sr)

cleaned_pet_df = cleaned_pet_df.assign(breed = sr)
```

```python
cleaned_pet_df= cleaned_pet_df.merge(states_csv, left_on='state',right_on='State', how='left')
```

```python
#formatting
cleaned_pet_df=cleaned_pet_df.drop(columns=['State', 'Latitude','Longitude'])
cleaned_pet_df=cleaned_pet_df.rename(columns={"City": "state_name"})
cleaned_pet_df
```

```python
#Establish a connection to SQL: 
from sqlalchemy import create_engine

# Creating connection object
protocol = 'postgresql'
username = 'postgres'
password = '2022'
host = 'localhost'
port = 5432
database_name = 'petfriends'
```

```python
#Create engine connection:
connection_string = f'{protocol}://{username}:{password}@{host}:{port}/{database_name}'

engine = create_engine(connection_string)
```

```python
cleaned_pet_df.to_sql(name='adoptees', con=engine, if_exists='replace', index=False)
friends = engine.execute("SELECT name, age, gender, city, breed, state FROM adoptees ORDER BY state ASC").fetchall()
```
