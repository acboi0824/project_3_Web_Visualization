```python
import pandas as pd

cleaned_pet_df = pd.read_csv('/Users/kyndallbutler/GTECH/project_3_Web_Visualization/data/pet_data_cleaned_more_data.csv')
cleaned_pet_df.head()
```

```python
sr = cleaned_pet_df['primary']
print(sr)

cleaned_pet_df = cleaned_pet_df.assign(breed = sr)
```

```python
#Establish a connection to SQL: 
from sqlalchemy import create_engine

# Creating connection object
protocol = 'postgresql'
username = 'postgres'
password = 'postgre'
host = 'localhost'
port = 5433
database_name = 'petfriends'
```

```python
#Create engine connection:
connection_string = f'{protocol}://{username}:{password}@{host}:{port}/{database_name}'

engine = create_engine(connection_string)
```

```python
cleaned_pet_df.to_sql(name='adoptees', con=engine, if_exists='replace', index=False)
friends = engine.execute("SELECT id, url, species, type, name, age, gender, zip, status, large_photo, medium_photo, city, breed, state FROM adoptees ORDER BY state ASC").fetchall()
```

```python
#Creating a new dataframe for adoptable pets:
adoptables = cleaned_pet_df[['id', 'type', 'name', 'age', 'gender', 'city', 'breed', 'city', 'zip', 'large_photo', 
                             'state', 'species', 'status', 'url']]
i = adoptables[(adoptables.status == 'adopted')].index

#Dropping adopted pets and exporting available pets to csv:
avail = adoptables.drop(i)
avail.head(n=25)
```
