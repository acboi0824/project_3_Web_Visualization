```python
import pandas as pd

cleaned_pet_df = pd.read_csv('pet_data_cleaned.csv')
print(cleaned_pet_df)
```

```python
sr = cleaned_pet_df['primary']
print(sr)

cleaned_pet_df = cleaned_pet_df.assign(breed = sr)
```

```python
#Establish a connection to SQL: 
from sqlalchemy import create_engine

# Creating connection object
protocol = 'postgresql'
username = 'postgres'
password = 'postgre'
host = 'localhost'
port = 5433
database_name = 'petfriends'
```

```python
#Create engine connection:
connection_string = f'{protocol}://{username}:{password}@{host}:{port}/{database_name}'

engine = create_engine(connection_string)
```

```python
cleaned_pet_df.to_sql(name='adoptees', con=engine, if_exists='replace', index=False)
friends = engine.execute("SELECT type, name, age, gender, city, breed, state FROM adoptees ORDER BY state ASC").fetchall()
```

```python
adoptables = cleaned_pet_df[['id', 'type', 'name', 'age', 'gender', 'city', 'breed', 'state', 'status']]
avail = adoptables.set_index("status")
avail.head()
```

```python
available = avail.drop("adopted")
available.head(n=25)
```
